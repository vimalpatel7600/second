```python
value = input('write an alpha : ')

if value== 'a' or value== 'A' :
    for row in range(5):
        for col in range(5):
            if row==0 or row==2 or col==0 or col==4:
                print('*',end=' ')
            else :
                print(' ',end=' ')
        print()

if value== 'b' or value== 'B' :

    for row in range(5):
        for col in range(5):
            if row==0 or row==2 or col==0 or col==4 or row==4:
                print('*',end=' ')
            else :
                print(' ',end=' ')
        print()
        
if value== 'c' or value== 'C' :
    
    for row in range(5):
        for col in range(5):
            if row==0 or row==4 or col==0 :
                print('*',end=' ')
            else :
                print(' ',end=' ')
        print()
        
if value== 'd' or value== 'D' :

    for row in range(5):
        for col in range(5):
            if row==0 or row==4 or col==0 or col==4 :
                print('*',end=' ')
            else :
                print(' ',end=' ')
    
        print()
        
if value== 'e' or value== 'E' :

    for row in range(5):
        for col in range(5):
            if row==0 or row==4 or col==0 or row==2 :
                print('*',end=' ')
            else :
                print(' ',end=' ')
    
        print()
        
if value== 'f' or value== 'f' :

    for row in range(5):
        for col in range(5):
            if row==0 or row==2 or col==0  :
                print('*',end=' ')
            else :
                print(' ',end=' ')
    
        print()
        
if value== 'g' or value== 'G' :

    for row in range(5):
        for col in range(6):
            if (row==0 and col<5) or (row==4 and col<4) or col==0 or (col==3 and row>1) or (row==2 and col>2) or (col==5 and row>1):
                print('*',end=' ')
            else :
                print(' ',end=' ')
    
        print()
        
if value== 'h' or value== 'H' :

    for row in range(5):
        for col in range(5):
            if col==0 or col==4 or row==2 :
                print('*',end=' ')
            else :
                print(' ',end=' ')
    
        print()
        
if value== 'i' or value== 'I' :

    for row in range(5):
        for col in range(5):
            if row==0 or row==4 or col==2 :
                print('*',end=' ')
            else :
                print(' ',end=' ')
    
        print()
        
if value== 'j' or value== 'J' :

    for row in range(5):
        for col in range(5):
            if (row==4 and col<3) or col==2 or row==0 :
                print('*',end=' ')
            else :
                print(' ',end=' ')
    
        print()
        
if value== 'k' or value== 'K' :

    for row in range(5):
        for col in range(5):
            if col==0 or (row-col==1 and row>1) or (row+col==3 and row<3):
                print('*',end=' ')
            else :
                print(' ',end=' ')
        print()
        
if value== 'l' or value== 'L' :

    for row in range(5):
        for col in range(4):
            if row==4 or col==0 :
                print('*',end=' ')
            else :
                print(' ',end=' ')
    
        print()
        
if value== 'm' or value== 'M' :

    for row in range(5):
        for col in range(5):
            if col==4 or col==0 or (row-col==0 and row<3 ) or (row+col==4 and row<3):
                print('*',end=' ')
            else :
                print(' ',end=' ')
    
        print()
        
if value== 'n' or value== 'N' :

    for row in range(5):
        for col in range(5):
            if col==0 or col==4 or row-col==0 :
                print('*',end=' ')
            else :
                print(' ',end=' ')
    
        print()
        
if value== 'o' or value== 'O' :

    for row in range(5):
        for col in range(5):
            if row==0 or row==4 or col==0 or col==4 :
                print('*',end=' ')
            else :
                print(' ',end=' ')
    
        print()
        
if value== 'p' or value== 'P' :

    for row in range(5):
        for col in range(5):
            if row==0 or row==2 or col==0 or (col==4 and row<3) :
                print('*',end=' ')
            else :
                print(' ',end=' ')
    
        print()

if value== 'q' or value== 'Q' :

    for row in range(6):
        for col in range(6):
            if (row==0 and col<5) or (row==4 and col<4) or (col==0 and row<5) or (col==4 and row<5) or (row-col==0 and col>2) :
                print('*',end=' ')
            else :
                print(' ',end=' ')
    
        print()
        
if value== 'r' or value== 'R' :

    for row in range(5):
        for col in range(5):
            if row==0 or row==2 or col==0 or (col==4 and row<3) or (row-col==0 and row>1):
                print('*',end=' ')
            else :
                print(' ',end=' ')
    
        print()
        
if value== 's' or value== 'S' :

    for row in range(5):
        for col in range(5):
            if row==0 or row==4 or (col==0 and row<3) or row==2 or (col==4 and row>1) :
                print('*',end=' ')
            else :
                print(' ',end=' ')
    
        print()
        
if value== 't' or value== 'T' :

    for row in range(5):
        for col in range(5):
            if row==0 or col==2 :
                print('*',end=' ')
            else :
                print(' ',end=' ')
    
        print()
        
if value== 'u' or value== 'U' :

    for row in range(5):
        for col in range(5):
            if row==4 or col==4 or col==0 :
                print('*',end=' ')
            else :
                print(' ',end=' ')
    
        print()
        
if value== 'v' or value== 'V' :

    for row in range(5):
        for col in range(5):
            if (col==0 and row<3) or (col==4 and row<3) or (row-col==2) or (row+col==6) :
                print('*',end=' ')
            else :
                print(' ',end=' ')
    
        print()
        
if value== 'v' or value== 'V' :

    for row in range(5):
        for col in range(5):
            if (col==0 and row<3) or (col==4 and row<3) or (row-col==2) or (row+col==6) :
                print('*',end=' ')
            else :
                print(' ',end=' ')
    
        print()
        
if value== 'x' or value== 'X' :

    for row in range(5):
        for col in range(5):
            if row-col==0 or row+col==4:
                print('*',end=' ')
            else :
                print(' ',end=' ')
    
        print()
        
if value== 'y' or value== 'Y' :

    for row in range(5):
        for col in range(5):
            if (row-col==0 and row<3) or (row+col==4 and row<3) or (col==2 and row>1) :
                print('*',end=' ')
            else :
                print(' ',end=' ')
    
        print()
        
if value== 'z' or value== 'Z' :

    for row in range(5):
        for col in range(5):
            if row==0 or row==4 or row+col==4 :
                print('*',end=' ')
            else :
                print(' ',end=' ')
    
        print()
```
